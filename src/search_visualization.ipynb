```python
import os

import matplotlib.pyplot as plt

from utils import load_img
from image_retrieval import retrieve_imgs
```

```python
def show_first_imgs(img_name, img_dir, dataset_dir, dataset, feature_extractor):
    fig, axes = plt.subplots(3, 5, figsize=(14, 6))

    query_img = load_img(os.path.join(img_dir, img_name))
    axes[0, 2].imshow(query_img)
    axes[0, 2].set_title("Query Image")

    query_result = retrieve_imgs(img_name, img_dir, dataset, feature_extractor)
    query_result = query_result[:10]

    for i, ax in enumerate(axes.flat):
        ax.set_axis_off()

    for i, ax in enumerate(axes.flat[5:]):
        img = load_img(os.path.join(dataset_dir, query_result[i]))
        ax.imshow(img)

    plt.show()
```

```python
show_first_imgs(
    '0_8c54ccf63a12ee74.jpg',
    '../datasets/GPR1200_dataset/',
    '../datasets/GPR1200_dataset/',
    'GPR1200',
    'cnn')
```
